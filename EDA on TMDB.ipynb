# Preprocess Data

```python
# To replace Nans to None values from all columns
train_df.fillna(value='',method=None, inplace = True)
test_df.fillna(value='',method=None, inplace = True)
```

```python
# To get genres into a list of list
dict_cols = ['belongs_to_collection','genres', 'production_companies', 'production_countries', 'spoken_languages','Keywords','cast','crew']
keys_dict = {}
for cols in dict_cols:
    keys_dict[cols] = list(eval(train_df[cols][0])[0].keys())
    print('Keys in "{}"  : {}'.format(cols,keys_dict[cols]))
```

```python
def get_ls_ls_from_dict(row, key):
    ls_genres = []
    ls = eval(row)
    cl_ls = [x for x in ls if x]
    for val in cl_ls:
        ls_genres.append(val[key])
    return ls_genres
```

```python
dummy_train_df = train_df.head(10).copy()
```

```python
dummy_train_df['ls_genres'] = dummy_train_df['genres'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['genres'][1]) if x else None)
dummy_train_df['ls_belongs_to_collection'] = dummy_train_df['belongs_to_collection'].apply(lambda x: get_ls_ls_from_dict(x,keys_dict['belongs_to_collection'][1]) if x else None)
dummy_train_df['ls_production_companies'] = dummy_train_df['production_companies'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['production_companies'][0]) if x else None)
dummy_train_df['ls_production_countries'] = dummy_train_df['production_countries'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['production_countries'][0]) if x else None)
dummy_train_df['ls_spoken_languages'] = dummy_train_df['spoken_languages'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['spoken_languages'][0]) if x else None)
dummy_train_df['ls_Keywords'] = dummy_train_df['Keywords'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['Keywords'][1]) if x else None)
dummy_train_df['ls_cast_name'] = dummy_train_df['cast'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['cast'][5]) if x else None)
dummy_train_df['ls_cast_gender'] = dummy_train_df['cast'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['cast'][3]) if x else None)
dummy_train_df['ls_crew_dept'] = dummy_train_df['crew'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['crew'][1]) if x else None)
dummy_train_df['ls_crew_gender'] = dummy_train_df['crew'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['crew'][2]) if x else None)
dummy_train_df['ls_crew_job'] = dummy_train_df['crew'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['crew'][4]) if x else None)
dummy_train_df['ls_crew_name'] = dummy_train_df['crew'].apply(lambda x: get_ls_ls_from_dict(x, keys_dict['crew'][5]) if x else None)
```
